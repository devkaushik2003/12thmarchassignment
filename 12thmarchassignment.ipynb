Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import numpy as np
import scipy.stats as stat
sample_mean = 50
sample_std = 5
n = 100
ci = 95
significance_value = 0.05
dof = n-1

t_value = stat.t.ppf(1-significance_value/2,n-1)

margin_of_error = t_value*(sample_std/(n**0.5))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print([lower_bound,upper_bound])
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
observed_data = [20,12,12,15,11.5,15]
expected_data = [0.2*95 , 0.2 *95 , 0.2*95 , 0.1*95,0.1*95,0.1*95]
chisquare_stats , p_value = stat.chisquare(observed_data,expected_data)
significance_value = 0.05
dof = len(expected_data) - 1
critical_value  = stat.chi2.ppf(1-significance_value,dof)
if(chisquare_stats> critical_value):
    print(" reject null hypothesis")
else:
    print("we failed to reject null hypothesis")

```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
from statsmodels.stats.proportion import proportion_confint

n = 500 # sample size
p = 60/n # sample proportion
alpha = 0.05 # significance level

lower_ci, upper_ci = proportion_confint(count=p*n, nobs=n, alpha=alpha, method='wilson')
print(f"95% Confidence Interval: ({lower_ci:.4f}, {upper_ci:.4f})")


```

Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the
area corresponding to a chi-square statistic of 15.

so in this question first we took the data by using np.linspace function and then using stat.chi2.pdf we get probability density function so that we get the values for y axis

```python
import numpy as np
import matplotlib.pyplot as plt
dof = 10
#plot the graph
x = np.linspace(0,30,100)
y = stat.chi2.pdf(x,dof)
fig , ax = plt.subplots()
ax.plot(x,y , "r-" , linewidth = 3)



x_shade = np.linspace(15,30,50)
y_shade = stat.chi2.pdf(x_shade,dof)
ax.fill_between(x_shade,y_shade,color = "blue",alpha = 0.5)

ax.set_xlabel('Chi-square statistic')
ax.set_ylabel('Probability density')
plt.show()
```

Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe
45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the
expected frequencies of a fair coin. Use a significance level of 0.05.

```python
observed_data = [55,45]
expected_data = [50,50]
chisquarestats , pvalue  = stat.chisquare(observed_data,expected_data)

dof = len(expected_data) -1
significance_value = 0.05
critical_value = stat.chi2.ppf(1-significance_value,dof)
if(chisquarestats > critical_value):
    print("reject null")
else:
    print("fail to reject null")
```
